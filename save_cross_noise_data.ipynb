```python
def wgn(x, snr):
    
    snr = 10**(snr/10.0)
    xpower = np.sum(x**2)/len(x)
    npower = xpower / snr
    
    return np.random.randn(len(x)) * np.sqrt(npower)


def add_noise(data):
    
    rand_data = wgn(data, snr_num)
    
    data = data + rand_data
    
    return data
```
